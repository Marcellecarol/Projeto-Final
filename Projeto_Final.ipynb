````python
# Instalação das bibliotecas necessárias
!pip install --upgrade google-cloud-bigquery pandas matplotlib seaborn

# Importando as bibliotecas
from google.cloud import bigquery
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Defina seu ID do projeto
project_id = 'silken-waters-434723-d9'

# Inicialize o cliente BigQuery
client = bigquery.Client(project=project_id)

# Defina a consulta SQL
query = """
SELECT *
FROM `basedosdados.br_ms_sia.producao_ambulatorial`
LIMIT 1000
"""

# Execute a consulta e armazene o resultado em um DataFrame
df = client.query(query).to_dataframe()

# Exibir as primeiras linhas do DataFrame
print("Primeiras linhas do DataFrame:")
print(df.head())

# Exibir os nomes das colunas
print("\nNomes das colunas:")
print(df.columns.tolist())  # Mostra as colunas em uma lista

# Analisar a eficiência do sistema de saúde

# Visualização da eficiência do sistema de saúde (supondo que exista uma coluna chamada 'tempo_atendimento')
if 'tempo_atendimento' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.histplot(df['tempo_atendimento'], bins=30, kde=True, color='skyblue')
    plt.title('Distribuição do Tempo de Atendimento', fontsize=16)
    plt.xlabel('Tempo de Atendimento', fontsize=14)
    plt.ylabel('Frequência', fontsize=14)
    plt.grid()
    plt.tight_layout()
    plt.show()
else:
    print("\nColuna 'tempo_atendimento' não encontrada.")

# Explorar a distribuição demográfica dos atendimentos
if 'idade' in df.columns:
    plt.figure(figsize=(12, 6))
    sns.histplot(df['idade'], bins=30, kde=True, color='salmon')
    plt.title('Distribuição Etária dos Atendimentos', fontsize=16)
    plt.xlabel('Idade', fontsize=14)
    plt.ylabel('Frequência', fontsize=14)
    plt.grid()
    plt.tight_layout()
    plt.show()
else:
    print("\nColuna 'idade' não encontrada.")

# Avaliar a evolução dos atendimentos
if 'data_atendimento' in df.columns:
    df['data_atendimento'] = pd.to_datetime(df['data_atendimento'])
    atendimentos_por_data = df['data_atendimento'].value_counts().sort_index()

    plt.figure(figsize=(12, 6))
    plt.plot(atendimentos_por_data.index, atendimentos_por_data.values, marker='o', linestyle='-', color='purple')
    plt.title('Evolução dos Atendimentos ao Longo do Tempo', fontsize=16)
    plt.xlabel('Data', fontsize=14)
    plt.ylabel('Número de Atendimentos', fontsize=14)
    plt.grid()
    plt.tight_layout()
    plt.show()
else:
    print("\nColuna 'data_atendimento' não encontrada.")

# Identificar os procedimentos mais comuns
if 'tipo_procedimento' in df.columns:
    procedimento_counts = df['tipo_procedimento'].value_counts().sort_values(ascending=False)

    plt.figure(figsize=(12, 6))
    sns.barplot(x=procedimento_counts.index, y=procedimento_counts.values, palette='viridis')
    plt.title('Procedimentos Mais Comuns', fontsize=16)
    plt.xlabel('Tipo de Procedimento', fontsize=14)
    plt.ylabel('Contagem', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
else:
    print("\nColuna 'tipo_procedimento' não encontrada.")
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Criando um DataFrame de exemplo (substitua isso pela sua entrada de dados)
data = {
    'valor_produzido_procedimento': [1000, 2000, 0, 3000, 4000],  # Adicionado valor zero
    'quantidade_produzida_procedimento': [10, 20, 0, 0, 50]  # Adicionado quantidade zero
}
procedimentos_df = pd.DataFrame(data)

# Criar um DataFrame com os detalhes dos procedimentos
detalhes_procedimentos = {
    'id_procedimento': [1, 2, 3, 4, 5],
    'descricao_procedimento': ['Consulta Médica', 'Exame de Sangue', 'Exame de Raio-X', 'Ultrassonografia', 'Internação Hospitalar']
}
detalhes_df = pd.DataFrame(detalhes_procedimentos)

# Adicionando as descrições de procedimentos ao DataFrame de procedimentos
procedimentos_df['descricao_procedimento'] = detalhes_df['descricao_procedimento'].values

# Criar um gráfico de barras
plt.figure(figsize=(10, 6))

# Plotar valor produzido
plt.bar(procedimentos_df.index - 0.2, procedimentos_df['valor_produzido_procedimento'], width=0.4, label='Valor Produzido', color='blue')

# Plotar quantidade produzida
plt.bar(procedimentos_df.index + 0.2, procedimentos_df['quantidade_produzida_procedimento'], width=0.4, label='Quantidade Produzida', color='orange')

# Adicionando título e rótulos
plt.title('Análise da Eficiência do Sistema de Saúde')
plt.xlabel('Procedimentos')
plt.ylabel('Valores')
plt.xticks(procedimentos_df.index, procedimentos_df['descricao_procedimento'])

# Ajustar limites do eixo Y
max_value = procedimentos_df[['valor_produzido_procedimento', 'quantidade_produzida_procedimento']].max().max() * 1.1
plt.ylim(0, max_value)

# Adicionando legendas
plt.legend()

# Adicionando rótulos de valor em cima das barras
for index, row in procedimentos_df.iterrows():
    plt.text(index - 0.2, row['valor_produzido_procedimento'] + 50, f'R$ {row["valor_produzido_procedimento"]}', ha='center')
    plt.text(index + 0.2, row['quantidade_produzida_procedimento'] + 1, str(row['quantidade_produzida_procedimento']), ha='center')

# Exibir o gráfico
plt.grid(axis='y')
plt.tight_layout()
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Criando um DataFrame de exemplo (substitua isso pela sua entrada de dados)
data = {
    'idade_paciente': [25, 30, 35, 40, 22, 30, 40, 55, 60, 45, 50, 35, 28, 33, 36, 29, 31, 42, 38, 47]
}
df = pd.DataFrame(data)

# Configurando o estilo do seaborn
sns.set(style='whitegrid')

# Gráfico 2: Distribuição demográfica dos atendimentos
plt.figure(figsize=(12, 6))
sns.histplot(df['idade_paciente'], bins=30, kde=True, color='skyblue', edgecolor='black')

# Adicionando título e rótulos
plt.title('Distribuição Demográfica dos Atendimentos por Idade', fontsize=18, fontweight='bold')
plt.xlabel('Idade do Paciente', fontsize=14)
plt.ylabel('Frequência', fontsize=14)

# Linha da média da idade
mean_age = df['idade_paciente'].mean()
plt.axvline(mean_age, color='red', linestyle='--', label='Idade Média')

# Anotando a média
plt.text(mean_age + 1, 2, f'Média: {mean_age:.1f}', color='red', fontsize=12, ha='left')

# Melhorando os ticks dos eixos
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Legenda e grid
plt.legend(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Linhas de grade no eixo Y

# Ajustando o layout
plt.tight_layout()
plt.show()
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Simulação de dados agrupados para demonstração (substitua isso pelos seus dados reais)
import pandas as pd
import numpy as np
df_grouped = pd.DataFrame({
    'mes': np.tile(range(1, 13), 3),
    'ano': np.repeat([2022, 2023, 2024], 12),
    'atendimentos': np.random.randint(50, 200, 36)
})

# Configurações estéticas
sns.set(style='whitegrid')

# Criar o gráfico com melhorias visuais
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_grouped, x='mes', y='atendimentos', hue='ano', marker='o', palette='coolwarm', linewidth=2.5, markersize=10)

# Melhorias visuais
plt.title('Evolução dos Atendimentos ao Longo dos Anos', fontsize=20, fontweight='bold', color='darkblue')
plt.xlabel('Mês', fontsize=16)
plt.ylabel('Número de Atendimentos', fontsize=16)
plt.xticks(range(1, 13), ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'], fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='Ano', title_fontsize=14, fontsize=12, loc='upper left')

# Adicionar rótulos de dados
for index, row in df_grouped.iterrows():
    plt.text(row['mes'], row['atendimentos'] + 2, str(row['atendimentos']), color='black', ha='center')

# Ajuste final do gráfico
plt.grid(axis='both', linestyle='--', alpha=0.6)  # Linhas de grade suaves
plt.tight_layout()

# Exibir o gráfico
plt.show()
````
